## Training Data
Training data yang dilakukan menggunakan data yang telah diberikan sejumlah 138 Data yang memiliki resolusi 992x1403 pixel (0,707 ratio). Data ini kemudian di di kecilkan (rescale down) dengan ratio yang tidak merusak gambar asli menjadi 640x896 pixel (0,714 ratio) untuk kebutuhan training data neural network. Setelah itu data ini kemudian di rescale down lagi sebelum di training menggunakan AutoEncoder NN menjadi 160x224 pixel dengan ratio yang sama seperti 640x896 hanya diperkecil 4 kali lipat.

```python
# Memanggil Library yang dibutuhkan
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

import cv2
import h5py
import numpy as np
import tensorflow as tf

import keras
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D
from keras.models import Model
from keras.datasets import mnist
from keras.callbacks import TensorBoard
from keras import backend as K
from keras.models import Sequential
from keras.models import load_model
from keras.backend.tensorflow_backend import set_session

import skimage
from skimage import io, color, data, exposure, feature,  filters
from skimage.filters import rank
from skimage import future, graph, draw, measure, morphology
from skimage import segmentation, transform, restoration, util

from scipy import ndimage as ndi
```

```python
# Mendefinisikan Fungsi
def set_gpu_mem_alloc(mem_use):
    avail  = 4041
    percent = mem_use / avail
    config = tf.ConfigProto()
    config.gpu_options.per_process_gpu_memory_fraction = percent
    config.gpu_options.visible_device_list = "0"
    set_session(tf.Session(config=config))

def set_dpi_val(val):
    rcVal = val
    plt.rcParams['image.cmap'] = 'gray'
    plt.rcParams['savefig.dpi'] = rcVal
    plt.rcParams['figure.dpi'] = rcVal


def show_images(images, cols = 1, titles = None):
    """Display a list of images in a single figure with matplotlib.
    
    Parameters
    ---------
    images: List of np.arrays compatible with plt.imshow.
    
    cols (Default = 1): Number of columns in figure (number of rows is 
                        set to np.ceil(n_images/float(cols))).
    
    titles: List of titles corresponding to each image. Must have
            the same length as titles.
    """
    assert((titles is None)or (len(images) == len(titles)))
    n_images = len(images)
    if titles is None: titles = ['Image (%d)' % i for i in range(1,n_images + 1)]
    fig = plt.figure()
    for n, (image, title) in enumerate(zip(images, titles)):
        a = fig.add_subplot(cols, np.ceil(n_images/float(cols)), n + 1)
        if image.ndim == 2:
            plt.gray()
        plt.imshow(image)
        a.set_title(title)
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_images)
    plt.show()

def get_roi(image, region_size=80):
    #prepare image 
    gray = color.rgb2gray(image)
    thresh = filters.threshold_otsu(gray)
    image_th = gray < thresh
    
    bw = morphology.closing(image_th, morphology.square(3))
    
    # # remove artifacts connected to image border
    cleared = segmentation.clear_border(bw)
   
    # # label image regions
    label_image = measure.label(cleared)
    image_label_overlay = color.label2rgb(label_image, image=image)
    
    #showing data
    comb = [image, image_label_overlay]
    show_images(comb)
    
    ## get region of interest
    roi = []
    for region in measure.regionprops(label_image):
        # take regions with large enough areas
        if region.area >= region_size:
            # draw rectangle around segmented coins
            minr, minc, maxr, maxc = region.bbox
            roi.append(region.bbox)
    
    return roi

            
def normalize_to_uint8(image):
    # Normalize
    img = image * 255
    image = img.astype('uint8')
    return image

def create_white_image(height, width):
    img = np.zeros([height, width, 3],dtype=np.uint8)
    img.fill(255)
    return img

def replace_to_center(src_image, dst_image):
    # get shape
    bh, bw, _ = dst_image.shape
    sh, sw, _ = src_image.shape

    # get center
    lh = int((bh - sh) / 2 )  
    lw = int((bw - sw) / 2) 
    rh = bh - lh
    rw = bw - lw

    #normalize dimension error if dimension is not match
    new_img = dst_image
    nh, nw, _ = dst_image[lh:rh, lw:rw].shape

    #find difference dimension between new shape and replacement image
    kh = nh - sh
    kw = nw - sw

    #normalize dimension dimension
    rh = rh - kh
    rw = rw - kw

    #Replace Image in center
    new_img[lh:rh, lw:rw] = src_image
    return new_img


def reshape_images(images, dim):
    ln, _, _, d = images.shape
    h, w = dim
    
    r_img = images.astype('float32') 
    result = np.reshape(r_img, (ln, h, w, d))
    return result

def imread_convert(f):
    return imread(f).astype(np.uint8)

def rescale(images, dim):
    h, w = dim
    out = []
    ln = len(images)
    for i in range(0, ln):
        image = images[i]
        image_resized = transform.resize(image, (h, w))
        out.append(image_resized)
    out = np.array(out)
    
    info = str(h)+'x'+str(w)
    print ('Rescaled ',len(out),' Images with Dimension '+info)
    return out


def rescale_up(image):
    h, w = 896,640
    image_resized = transform.resize(image, (h, w))
    out = np.array(image_resized)
    return out

def rescale_size(image, dim):
    h,w = dim
    resized = transform.resize(image, (h, w))
    return np.array(resized)



def load_images(path, dim, is_limit=False, limit=None):
    # path in = data/training/asl/*_asl.jpg
    images = io.imread_collection(path)
    out = []
    c = 0
    for img in images:
        if is_limit:
            if c == limit:
                break
        img = rescale_size(img, dim)
        out.append(img)
        c = c + 1
    out = np.array(out)
    
    h, w=dim
    info = str(h)+'x'+str(w)
    print ('Loaded ',len(out),' Images with Dimension '+info)
    return out


def load_neural_network(path):
    model = load_model('saved/mask_model.h5')
    model.summary()
    print('Total Layer ',len(model.layers))
    return model

def test_single_data(model, image):
    h, w, d = 224, 160, 3
    dim = (h, w)
    single = np.reshape(image, (1, h, w, d))
    result = model.predict(single, batch_size=1)
    return result[0]

def test_multiple_data(image):
    h,w,d = 224,160,3
    single = np.reshape(image, (1, h, w, d))
    result = model.predict(single, batch_size=1)
    return result


def get_crop_roi_images(image, roi):
    cropped = []
    for r in roi:
        minr, minc, maxr, maxc = r
        img = image[minr:maxr,minc:maxc]
        img = normalize_to_uint8(img)
        cropped.append(img)
        plt.imshow(img)
        plt.show();
    return cropped
```

```python
%%time
# Mendefinisikan DPI dan penggunakan GPU
set_dpi_val(180)
set_gpu_mem_alloc(2000)
```

```python
%%time
path = 'data/training/asl/*_asl.jpg'

std_dim = (896, 640)
images = load_images(path, std_dim, True, 10)

nn_dim = (224,160)
rescale_images = rescale(images, nn_dim)

reshaped_images = reshape_images(rescale_images, nn_dim)
```

```python
%%time
nn_path = 'saved/mask_model.h5'
model = load_neural_network(nn_path)
```

```python
%%time
im_num = 0
nn_image = reshaped_images[im_num]
ori_image = images[im_num]
print('Shape ',nn_image.shape)
print('Type ',type(nn_image))

combined = [ori_image, nn_image]
show_images(combined)
```

```python
%%time
result = test_single_data(model, nn_image)
```

```python
result_image = normalize_to_uint8(result)
result_image = rescale_up(result_image)

combined = [ori_image, result_image]
show_images(combined)
```

```python
%%time
minimum_size_bbox = 100
roi = get_roi(result_image, minimum_size_bbox)    
```

```python
cropped_images = get_crop_roi_images(ori_image, roi)
```

```python
data = []
for img in cropped_images:
    white_image = create_white_image(112, 640)
    new_img = replace_to_center(img, white_image)
    data.append(new_img)
    plt.imshow(new_img)
    plt.show();
    new_img = None
```
